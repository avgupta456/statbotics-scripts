```python
from collections import defaultdict

import pandas as pd
import statbotics

sb = statbotics.Statbotics()
```

```python
keys = ["2022carv", "2022gal", "2022new", "2022roe", "2022tur", "2022hop"]
# keys = ["2022gal", "2022new", "2022roe", "2022tur"]
matches = []
for key in keys:
    matches.extend(sb.get_matches(event=key))
matches_df = pd.DataFrame(matches)
```

```python
teams_df = pd.DataFrame(sb.get_team_years(year=2022, limit=10000))
elos = defaultdict(lambda: 1500)
oprs = defaultdict(lambda: 13)
for _, r in teams_df.iterrows():
    elos[r.team] = r.elo_end
    oprs[r.team] = r.opr_no_fouls
```

```python
new_matches = []
for _, match in matches_df.iterrows():
    red_teams = [int(x) for x in match.red.split(",")]
    blue_teams = [int(x) for x in match.blue.split(",")]
    red_opr = sum([oprs[r] for r in red_teams])
    blue_opr = sum([oprs[b] for b in blue_teams])
    red_elo = sum([elos[r] for r in red_teams])
    blue_elo = sum([elos[b] for b in blue_teams])
    new_matches.append({
        "key": match.key,
        "red_teams": red_teams,
        "blue_teams": blue_teams,
        "red_opr": red_opr,
        "blue_opr": blue_opr,
        "red_elo": red_elo,
        "blue_elo": blue_elo,
        "max_opr": max(red_opr, blue_opr),
        "min_opr": min(red_opr, blue_opr),
        "sum_opr": red_opr + blue_opr,
        "diff_opr": abs(red_opr - blue_opr),
        "max_elo": max(red_elo, blue_elo),
        "min_elo": min(red_elo, blue_elo),
        "sum_elo": red_elo + blue_elo,
        "diff_elo": abs(red_elo - blue_elo),
    })
    
new_matches_df = pd.DataFrame(new_matches)
```
