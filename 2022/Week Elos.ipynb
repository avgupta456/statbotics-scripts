```python
import pandas as pd

import statbotics

sb = statbotics.Statbotics()
```

```python
team_years = sb.get_team_years(year=2022)
team_years_df = pd.DataFrame(team_years)
team_years_df = team_years_df[team_years_df.elo_max > 0]
```

```python
elo_df = team_years_df[["team", "name", "state", "country", "district", "elo_start", "elo_end", "elo_diff"]]

elo_df[elo_df.elo_end < 1900].sort_values(by="elo_diff", ascending=False)[:10]
```

```python
team_events_1 = sb.get_team_events(year=2022, week=1)
team_events_2 = sb.get_team_events(year=2022, week=2)

team_events_df_1 = pd.DataFrame(team_events_1)
team_events_df_2 = pd.DataFrame(team_events_2)

team_events_df_1 = team_events_df_1[team_events_df_1.elo_max > 0][["team", "team_name", "event", "state", "district", "elo_start", "elo_end", "wins", "losses", "ties"]]
team_events_df_2 = team_events_df_2[team_events_df_2.elo_max > 0][["team", "team_name", "event", "state", "district", "elo_start", "elo_end", "wins", "losses", "ties"]]
```

```python
display(team_events_df_1[team_events_df_1.elo_end > team_events_df_1.elo_start].sort_values(by="elo_end", ascending=False)[:25])
display(team_events_df_2[team_events_df_2.elo_end > team_events_df_2.elo_start].sort_values(by="elo_end", ascending=False)[:25])
```
